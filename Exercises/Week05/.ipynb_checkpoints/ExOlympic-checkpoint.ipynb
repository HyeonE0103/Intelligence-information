# 동계 올림픽 메달 개수 시각화하기
## 202100001234 홍길동

```python
import pandas as pd
```

```python
url = "https://ko.wikipedia.org/wiki/%EC%98%AC%EB%A6%BC%ED%94%BD_%EB%A9%94%EB%8B%AC_%EC%A7%91%EA%B3%84"
```

```python
# pandas 라이브러리로 URL을 활용하여 데이터 가져오기
df = pd.read_html(url)
df
```

```python
len(df)
```

```python
df[0]
```

```python
# 헤더를 열 이름(columns)으로, 나라 이름을 행 이름(index)로 지정
df = pd.read_html(url, header=0, index_col=0)
df[0]
```

```python
# 슬라이싱을 활용하여 하계올림픽에 대한 데이터 추출
a = df[0].iloc[:, :5]
a
```

```python
# 컬럼명 설정하기
a.columns = ['하계 참가 횟수', '금', '은', '동', '계']
a
```

```python
# 금메달 기준으로 내림차순 정렬하기
a = a.sort_values(by=['금'], ascending=False)
```

```python
# matplotlib 설정하기
import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.style.use("ggplot")
```

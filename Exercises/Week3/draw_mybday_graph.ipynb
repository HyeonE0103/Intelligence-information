```python
import csv
import matplotlib.pyplot as plt

f = open('incheon_weather.csv', 'r')
data = csv.reader(f)
next(data)
high = []
low = []

for row in data:
    if row[-1] != '' and row[-2] != '':
        if int(row[0].split('-')[0]) >= 1999:
            if row[0].split('-')[1] == '01' and row[0].split('-')[2] == '03':
                high.append(float(row[-1]))
                low.append(float(row[-2]))
    
f.close()

plt.rc('font', family = 'Malgun Gothic')
plt.title('내 생일 최고 & 최저 기온 변화 그래프(01월 03일)')

plt.rcParams['axes.unicode_minus'] = False
plt.plot(high, color='r', ls='--', marker='v', label='최고기온')
plt.plot(low, color='b', ls=':', marker='o', label='최저기온')

plt.legend(loc=4)
plt.show()
```

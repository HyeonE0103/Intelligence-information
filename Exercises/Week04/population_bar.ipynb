```python
import csv
import matplotlib.pyplot as plt

f = open('incheon_population.csv', 'r')
data = csv.reader(f)

male = []
female = []

name = input('인구 구조를 알고 싶은 동 이름 입력: ')

for row in data:
    if name in row[0]:
        for i in row[3:104]:
            male.append(-int(i))
        for i in row[106:207]:
            female.append(int(i))

plt.rc('font', family = 'Malgun Gothic')
plt.title(name + '연령별 인구수')

plt.style.use('ggplot') #격자 무늬 스타일 지정

plt.rcParams['axes.unicode_minus'] = False
plt.barh(range(101), male, label='남성')
plt.barh(range(101),female, label='여성')
plt.xlabel('연령')
plt.ylabel('인구수')

plt.legend()
plt.show()
```

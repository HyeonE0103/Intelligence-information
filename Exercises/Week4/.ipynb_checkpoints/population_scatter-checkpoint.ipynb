```python
import csv
import matplotlib.pyplot as plt
import math

f = open('incheon_population.csv', 'r')
data = csv.reader(f)

male, female = [], []
size = []
name = input('인구 구조를 알고 싶은 동 이름 입력: ')

for row in data:
    if name in row[0]:
        for i in range(3, 104):
            m, fm = int(row[i]), int(row[i+103])
            male.append(m)
            female.append(fm)
            size.append(math.sqrt(m+fm))
        break

plt.rc('font', family = 'Malgun Gothic')
plt.figure(figsize=(10,5), dpi=300)
plt.title(name + '연령별 인구수', fontsize=15, color='b')

plt.style.use('ggplot') #격자 무늬 스타일 지정

plt.scatter(male, female, c=range(101), alpha=0.8, cmap='jet') #컬러맵 적용
plt.colorbar()

max_val = max(max(male), max(female))
plt.axis([0, max_val, 0, max_val])
line = range(max_val)

plt.plot(line, line, color='g')
plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')

plt.show()
```
